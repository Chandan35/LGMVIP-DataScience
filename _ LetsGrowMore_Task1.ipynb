#### Importing libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

#### Loading Data 

```python
# saving filepath
file_path = 'IRIS.csv'
data = pd.read_csv(file_path) 
```
